```python
import cv2

def extract_frames(video_path, output_folder, num_frames):
    # Open the video file
    vidcap = cv2.VideoCapture(video_path)
    success, image = vidcap.read()
    count = 0

    # Calculate interval for evenly spaced frames
    total_frames = int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT))
    frame_interval = total_frames // num_frames

    # Extract frames
    while success and count < num_frames:
        # Write the current frame as an image
        cv2.imwrite(f"{output_folder}/frame_{count}.jpg", image)

        # Move to the next frame
        for _ in range(frame_interval):
            success, image = vidcap.read()
        count += 1

    # Release the VideoCapture object
    vidcap.release()
    print(f"{count} frames extracted successfully.")

# Example usage:
video_path = "/content/Simple Pendulum.mp4"
output_folder = "/content/drive/MyDrive/CODEV/Test_pendule"
num_frames = 25  # Change this to the number of frames you want to extract
extract_frames(video_path, output_folder, num_frames)
```
