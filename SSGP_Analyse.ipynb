```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import simps

# Remplacez 'your_file.csv' par le chemin de votre fichier CSV
file_path = 'mse_time_results.csv'

# Lire le fichier CSV
data = pd.read_csv(file_path)

# Vérifier les premières lignes du fichier pour confirmer la lecture correcte
print(data.head())

# Extraire les colonnes nécessaires
x = data.iloc[:, 1]  # Deuxième colonne
y = data.iloc[:, 2]  # Troisième colonne

# Tracer la troisième colonne en fonction de la deuxième
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', linestyle='-')

# Ajouter des titres et des étiquettes
plt.xlabel('Deuxième colonne')
plt.ylabel('Troisième colonne')
plt.title('Trace de la troisième colonne en fonction de la deuxième')

# Afficher la grille
plt.grid(True)

# Afficher le graphique
plt.show()
```

```python
import numpy as np
from scipy.integrate import simps

# Calculer l'intégrale de la courbe
integrale = simps(y, x)

# Calculer la période
periode = np.max(x) - np.min(x)

# Diviser l'intégrale par la période
resultat = integrale / periode

print(f"Intégrale de la courbe divisée par la période : {resultat}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps

# Exemple de données réelles (remplacez ces listes par vos propres données)
time_points = time_points  # Exemple de liste de temps régulièrement espacée
mse_values = mse_values  # Exemple de liste de MSE

# Vérifier que les deux listes ont la même longueur
assert len(time_points) == len(mse_values), "Les listes de temps et de MSE doivent avoir la même longueur."

# Tracer la courbe de la MSE en fonction du temps
plt.figure(figsize=(10, 6))
plt.plot(time_points, mse_values, marker='o', linestyle='-')

# Ajouter des titres et des étiquettes
plt.xlabel('Temps')
plt.ylabel('MSE')
plt.title('Évolution de la MSE en fonction du temps')

# Afficher la grille
plt.grid(True)

# Afficher le graphique
plt.show()

# Calculer l'intégrale de la courbe (aire sous la courbe)
integrale = simps(mse_values, time_points)

# Calculer la période (la différence entre la valeur maximale et minimale de time_points)
periode = np.max(time_points) - np.min(time_points)

# Diviser l'intégrale par la période
resultat = integrale / periode

print(f"Intégrale de la courbe divisée par la période : {resultat}")
```
